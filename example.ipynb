```python
import numpy as np
import matplotlib.pyplot as plt
from jacobi import jacobirot
np.random.seed(123)
```

```python
A = np.random.randn(10,10)
plt.imshow(A)
plt.colorbar()
```

```python
U,S,V = np.linalg.svd(A)
```

```python
n = A.shape[0]
for i in range(100):
    for i in range(n):
        for j in range(i+1,n):
            Jtheta, Jphi = jacobirot(A,i,j)
            A = Jtheta @ A @ Jphi
S0 = np.sort(np.abs(np.diag(A)))[::-1]
```

```python
err = np.abs(S - S0)
print(err)
```
